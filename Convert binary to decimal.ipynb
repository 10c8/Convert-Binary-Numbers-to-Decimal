```python
import numpy as np
import matplotlib.pyplot as plt
from keras.layers.normalization import BatchNormalization
from keras.layers.advanced_activations import LeakyReLU, PReLU
from keras.models import load_model
from keras import backend as K
from keras import regularizers
from keras.models import Sequential
from keras.layers import Dense, Activation,LSTM,SimpleRNN,Dropout
from keras.callbacks import History 
history = History()
```

```python
#return a dataset of shape=(shape0,shape1)
#each cell is 0 or 1
#shape0 is the number of instances
#shape1 is the size max of each instance
def create_x(shape0,shape1):
    dataset=np.array(())
    for i in range(0,shape0*shape1):
            dataset=np.insert(dataset,dataset.shape[0],np.random.randint(2))
    dataset=dataset.reshape((-1,shape1))
    return dataset

#return a dataset containing each x binary row to decimal
def create_y(x):
    y=np.array(())
    for i in x:
        number=''
        for j in i:
            number+=str(int(j))
        y=np.insert(y,y.shape[0],int(number,2))
    y=y.reshape((-1,1))
    return y
```

```python
def get_model(x_train,y_train,x_test,y_test,layer_size=10,batch_size=100,nb_epoch=10):
    model = Sequential()
    model.add(BatchNormalization(input_shape=(None, x_train.shape[2])))
    model.add(LSTM(units=layer_size,return_sequences=True,stateful=False,recurrent_dropout=0.2,
                   dropout=0.4,kernel_regularizer=regularizers.l2(0.00001)))
    model.add(LeakyReLU())
    model.add(BatchNormalization())
    model.add(LSTM(layer_size,return_sequences=False,stateful=False,recurrent_dropout=0.2,dropout=0.4,
                   kernel_regularizer=regularizers.l2(0.00001)))
    model.add(LeakyReLU())
    model.add(BatchNormalization())
    model.add(Dense(y_train.shape[1],activation="linear"))
    model.compile(loss='mse', optimizer='adam')
    model.fit(x_train,y_train, epochs=nb_epoch,batch_size=batch_size,validation_data=(x_test, y_test),verbose=2,
              shuffle=False,callbacks=[history])
    return model

#predict each sample of the test dataset
def predict_test(model,x_test,y_test):
    for i in range(0,x_test.shape[0]):
        a=x_test[i].reshape((1,x_test[i].shape[0],x_test[i].shape[1]))
        a=model.predict(a)[0][0]
        print 'predicted : ',round(a), 'true :',y_test[i,0], 'error :', abs(round(a)-y_test[i,0])

#predict on sample (ex:[0,1,0,0,1])
def unique_test(model,x):
    a=x.reshape((1,x.shape[0],1))
    a=model.predict(a)[0][0]
    print 'predicted : ',round(a), 'true :',y_test[i,0], 'error :', abs(a-y_test[i,0])
```

```python
x=create_x(10000,5)
x=x.reshape((-1,x.shape[1],1))
y=create_y(x)
```

```python
#ratio test/train
ratio_test_all=0.1

x_train=x[:-int(x.shape[0]*ratio_test_all),:,:]
y_train=y[:-int(y.shape[0]*ratio_test_all),:]
x_test=x[-int(x.shape[0]*ratio_test_all):,:,:]
y_test=y[-int(y.shape[0]*ratio_test_all):,:]

model=get_model(x_train,y_train,x_test,y_test,layer_size=10,batch_size=1000,nb_epoch=500)


#plot the curve of the error
"""plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.show()"""
```
